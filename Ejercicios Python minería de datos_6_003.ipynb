```python
#Se crea este dataframe para poder tener la siguiente tabla y poder hacer las gráficas

ARBM=ARB[ARB['IndicatorCode']=='SP.DYN.IMRT.IN'] 
ARBM=ARBM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
ARBM.rename(columns = {'Value':'ARB'}, inplace = True) 

MEAM=MEA[MEA['IndicatorCode']=='SP.DYN.IMRT.IN'] 
MEAM=MEAM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
MEAM.rename(columns = {'Value':'MEA'}, inplace = True) 

EASM=EAS[EAS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
EASM=EASM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
EASM.rename(columns = {'Value':'EAS'}, inplace = True) 

ECSM=ECS[ECS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
ECSM=ECSM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
ECSM.rename(columns = {'Value':'ECS'}, inplace = True) 


LCNM=LCN[LCN['IndicatorCode']=='SP.DYN.IMRT.IN'] 
LCNM=LCNM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
LCNM.rename(columns = {'Value':'LCN'}, inplace = True) 


NACM=NAC[NAC['IndicatorCode']=='SP.DYN.IMRT.IN'] 
NACM=NACM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
NACM.rename(columns = {'Value':'NAC'}, inplace = True) 


SASM=SAS[SAS['IndicatorCode']=='SP.DYN.IMRT.IN'] 
SASM=SASM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
SASM.rename(columns = {'Value':'SAS'}, inplace = True) 

SSFM=SSF[SSF['IndicatorCode']=='SP.DYN.IMRT.IN'] 
SSFM=SSFM.drop(['CountryName','IndicatorName','IndicatorCode','CountryCode'],axis=1)
SSFM.rename(columns = {'Value':'SSF'}, inplace = True) 

a = pd.merge(ARBM, MEAM, on='Year')
b = pd.merge(EASM, ECSM, on='Year')
c = pd.merge(LCNM, NACM, on='Year')
d = pd.merge(SASM, SSFM, on='Year')
A = pd.merge(a,b, on='Year')
B = pd.merge(c,d, on='Year')
result = pd.merge(A,B, on='Year')
result.to_csv('Tabla_Mortalidad_Filas_years_colum_pais.csv', index=False)

result
```

```python
fig,ax=plt.subplots()
#ARB,MEA,EAS,ECS,LCN,NAC,SAS,SSF
ax.plot(result['Year'], result['ARB'] , label="ARB".format(0))
ax.plot(result['Year'], result['MEA'] , label="MEA".format(1))
ax.plot(result['Year'], result['EAS'] , label="EAS".format(2))
ax.plot(result['Year'], result['LCN'] , label="LCN".format(3))
ax.plot(result['Year'], result['NAC'] , label="NAC".format(4))
ax.plot(result['Year'], result['SAS'] , label="SAS".format(5))
ax.plot(result['Year'], result['SSF'] , label="SSF".format(6))
ax.plot(result['Year'], result['ECS'] , label="ECS".format(7))
leg = ax.legend(loc="upper left", bbox_to_anchor=[0, 1],
                 ncol=2, shadow=True, title="CountryCode", fancybox=True)
leg.get_title().set_color("red")
```
